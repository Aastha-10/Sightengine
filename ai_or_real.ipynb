```python
import os
import requests
import json

folder_path = r"C:\Users\aastha\.vscode\extensions\sourcery.sourcery-1.37.0-win32-x64\New folder\AI_generated"
api_user = '1838034711'   
api_secret = 'TkPXwbZ6Bm7CDHc5Re8KEsRHgFxFSjjU' 
models = 'genai'

for filename in os.listdir(folder_path):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.webp', '.bmp', '.gif')):
        file_path = os.path.join(folder_path, filename)
        with open(file_path, 'rb') as image_file:
            files = {'media': image_file}
            params = {
                'models': models,
                'api_user': api_user,
                'api_secret': api_secret
            }
            response = requests.post('https://api.sightengine.com/1.0/check.json', files=files, data=params)
            output = json.loads(response.text)
            print(f"Results for {filename}:\n{json.dumps(output, indent=2)}\n")
```
